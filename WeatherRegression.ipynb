```python
import numpy as numpyinstace
import pandas as pandasinstace
import matplotlib.pyplot as matplotlibinstace
import seaborn as seaborninstace
```

```python
weather= pandasinstace.read_csv('weather.csv')
display(weather.info())
```

```python
weather.describe()
```

```python
seaborninstace.scatterplot(x='Temp9am',y='Humidity9am',data=weather)
```

```python
seaborninstace.jointplot(x='Temp9am',y='Humidity9am',data=weather,kind='hex')
```

```python
seaborninstace.lmplot(x='Temp3pm',y='Humidity3pm',data=weather)
```

```python
X_Features=weather[['Humidity3pm']]
X_Features
```

```python
Y_prediction=weather[['Temp9am']]
Y_prediction
```

```python
weather.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(5,5))
plt.tight_layout()
sns.distplot(weather['MaxTemp'])
```

```python
Y_prediction = weather[['MinTemp']]
print(Y_prediction)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_Features,Y_prediction)
temperaturePrediction=model.predict(X_Features)
print(temperaturePrediction)
```

```python
matplotlibinstace.scatter(Y_prediction,temperaturePrediction)
matplotlibinstace.xlabel('MinTemp')
matplotlibinstace.ylabel('MaxTemp')
matplotlibinstace.show()
```

```python
sns.distplot(Y_prediction-temperaturePrediction)
```

```python
from sklearn import metrics
metrics.mean_absolute_error(Y_prediction,temperaturePrediction)
```

```python
metrics.mean_squared_error(Y_prediction,temperaturePrediction)
```

```python
numpyinstace.sqrt(metrics.mean_squared_error(Y_prediction,temperaturePrediction))
```
